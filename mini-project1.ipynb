```python
#==========Estimate time series regressions, Rpt - Rft = a + βiM (RMt - Rft) + eit, for each of the 30 industry portfolios. 
#Perform the multivariate GRS test; 
#please report both the GRS F-statistic and its p-value. Use the market portfolio proxy RM-RF.


#creating portfolio excess return
VWreturns.iloc[:,:VWreturns.shape[1]] = VWreturns.iloc[:,:VWreturns.shape[1]].subtract(MKTRF['RF'],axis=0)
VWreturns.head()
```

```python
import statsmodels.api as sm
```

```python
#create Epsilons
Epsilons = VWreturns[:].iloc[:,0:30]*0
Epsilons.head()
```

```python
#create estimates
Estimates = pd.DataFrame(VWreturns.iloc[:,0:30].mean())
Estimates.columns=['alpha_ts']
Estimates.head()
```

```python
#do the regression analysis
x = sm.add_constant(MKTRF.iloc[:1069,0])
for portfolios in VWreturns.columns[0:30]:
    y= VWreturns.loc[:,portfolios]
    results = sm.OLS(y,x).fit()
    Estimates.set_value(portfolios,'alpha_ts',results.params[0])
    Epsilons[portfolios]=results.resid
```

```python
Sigma=Epsilons.cov() 
# sample size #即有多少个日期
T=Epsilons.shape[0]
# number of test assets
N=Epsilons.shape[1]
# sharpe ratio of the factor
SRmkt = MKTRF.iloc[:1069, 0].mean()/MKTRF.iloc[:1069, 0].std()#这里factor就是market，求market 的SR
```

```python
# GRS stats
Chi2 = T/(1+SRmkt**2)*(Estimates.alpha_ts.values @ np.linalg.inv(Sigma.values) @ Estimates.alpha_ts.values.T)
Chi2
```

```python
#F stats
F = T*(T-N-1)*Chi2/(N*(T-2))
F
```

```python
import scipy.stats as stats
1 - stats.chi2.cdf(Chi2, N)
```
