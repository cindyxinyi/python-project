 - Estimate βiM for each portfolio (denote the estimate biM). Assume that the betas do not change over time; hence, you can estimate the betas using full-period OLS regressions.

 - Each month estimate the cross-sectional regression, Rit = γ0t + γMt biM + nit. Note: the estimated beta (biM ) is the same for every time period (i.e., it has no time subscript).

 - Compute the time series average of the estimates of γ0 and γM. In addition, compute the standard error and t-stat of the time series averages.

```python
#=================estimate beta==============
Estimates = pd.DataFrame(Data.iloc[:,0:(Data.shape[1]-2)].mean())
Estimates.columns = ['AvgReturns']
Estimates['beta'] = Estimates['AvgReturns']*0

import statsmodels.api as sm

for portfolios in Estimates.index:
    x = sm.add_constant(Data['Mkt-RF'])
    y = Data[portfolios]
    results = sm.OLS(y, x, missing = 'drop').fit()
    Estimates.set_value(portfolios,'beta',results.params[1])

Estimates.head()
```

```python
#===========estimating gamma0,gamma_m,gamma_size,gamma_bm============
for i in range(1,Data2.shape[0]):
    y=(Data2.loc[i,:]).iloc[0:(Data2.shape[1]-2)]
    x1 = Estimates['beta']
    x2 = np.log((Merged2.loc[i,:]).iloc[0:(SIZE.shape[1]-1)].astype(np.float64))
    x3 = np.log((Merged2.loc[i,:]).iloc[(SIZE.shape[1]):].astype(np.float64))
    c = pd.DataFrame({'x1':np.array(x1),'x2':np.array(x2),'x3':np.array(x3)})
    x = sm.add_constant(c)
    xx = x.set_index(y.index)
    results = sm.OLS(y,xx,missing='drop').fit()  

    FamaMacbeth_f.set_value(i,'gamma0',results.params[0])#intercept
    FamaMacbeth_f.set_value(i,'gamma_m',results.params[1])
    FamaMacbeth_f.set_value(i,'gamma_size',results.params[2])
    FamaMacbeth_f.set_value(i,'gamma_bm',results.params[3])
```
